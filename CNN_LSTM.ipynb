```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import (
    Input, Conv1D, MaxPooling1D, BatchNormalization, Dropout,
    LSTM, Dense, Flatten, TimeDistributed, Reshape, GlobalAveragePooling1D
)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2

SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)


def load_uci_har(path='UCI_HAR_Dataset/'):
    def load_signals(split):
        signals_path = f"{path}/{split}/Inertial Signals/"
        signal_files = [
            'body_acc_x', 'body_acc_y', 'body_acc_z',
            'body_gyro_x', 'body_gyro_y', 'body_gyro_z',
            'total_acc_x', 'total_acc_y', 'total_acc_z'
        ]
        signals = []
        for sig in signal_files:
            data = np.loadtxt(f"{signals_path}/{sig}_{split}.txt")
            signals.append(data)
        return np.stack(signals, axis=2)

    def load_labels(split):
        labels = np.loadtxt(f"{path}/{split}/y_{split}.txt", dtype=int)
        return labels - 1

    try:
        X_train = load_signals('train')
        X_test  = load_signals('test')
        y_train = load_labels('train')
        y_test  = load_labels('test')
        print(f"[UCI HAR] Loaded successfully. Train: {X_train.shape}, Test: {X_test.shape}")
        return X_train, X_test, y_train, y_test
    except FileNotFoundError:
        print("[UCI HAR] Dataset not found — generating synthetic data for demo.")
        return generate_synthetic_har()


def generate_synthetic_har(n_train=7352, n_test=2947,
                            timesteps=128, n_features=9, n_classes=6):

    ACTIVITY_NAMES = [
        'Walking', 'Walking Upstairs', 'Walking Downstairs',
        'Sitting', 'Standing', 'Laying'
    ]
    t = np.linspace(0, 2 * np.pi, timesteps)

    def make_samples(n, seed_offset=0):
        np.random.seed(SEED + seed_offset)
        X = np.zeros((n, timesteps, n_features))
        y = np.random.randint(0, n_classes, n)
        freqs = [1.5, 2.0, 2.5, 0.3, 0.1, 0.05]

        for i in range(n):
            cls = y[i]
            f   = freqs[cls]
            for feat in range(n_features):
                phase = np.random.uniform(0, np.pi)
                amp   = np.random.uniform(0.7, 1.3)
                X[i, :, feat] = (amp * np.sin(f * t + phase)
                                 + 0.15 * np.random.randn(timesteps))
        return X, y

    X_train, y_train = make_samples(n_train, seed_offset=0)
    X_test,  y_test  = make_samples(n_test,  seed_offset=1)

    print(f"[Synthetic HAR] Train: {X_train.shape}, Test: {X_test.shape}")
    return X_train, X_test, y_train, y_test


def preprocess(X_train, X_test, y_train, y_test, n_classes=6):
    mean = X_train.mean(axis=(0, 1), keepdims=True)
    std  = X_train.std(axis=(0, 1), keepdims=True) + 1e-8
    X_train = (X_train - mean) / std
    X_test  = (X_test  - mean) / std

    Y_train = to_categorical(y_train, n_classes)
    Y_test  = to_categorical(y_test,  n_classes)

    return X_train, X_test, Y_train, Y_test, y_train, y_test


def build_cnn_lstm(input_shape, n_classes,
                   cnn_filters=(64, 128, 256),
                   lstm_units=(128, 64),
                   dropout_rate=0.4):

    inputs = Input(shape=input_shape, name='sensor_input')

    x = Conv1D(cnn_filters[0], 5, padding='same', activation='relu', kernel_regularizer=l2(1e-4))(inputs)
    x = BatchNormalization()(x)
    x = MaxPooling1D(2)(x)
    x = Dropout(dropout_rate / 2)(x)

    x = Conv1D(cnn_filters[1], 3, padding='same', activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = MaxPooling1D(2)(x)
    x = Dropout(dropout_rate / 2)(x)

    x = Conv1D(cnn_filters[2], 3, padding='same', activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Dropout(dropout_rate / 2)(x)

    x = LSTM(lstm_units[0], return_sequences=True, dropout=dropout_rate)(x)
    x = LSTM(lstm_units[1], return_sequences=False, dropout=dropout_rate)(x)

    x = Dense(128, activation='relu')(x)
    x = Dropout(dropout_rate)(x)
    outputs = Dense(n_classes, activation='softmax')(x)

    return Model(inputs, outputs)


def train_model(model, X_train, Y_train, X_test, Y_test):
    model.compile(optimizer=Adam(1e-3),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    history = model.fit(
        X_train, Y_train,
        validation_data=(X_test, Y_test),
        epochs=50,
        batch_size=64,
        verbose=1
    )
    return history


def evaluate_model(model, X_test, Y_test, y_test):
    loss, acc = model.evaluate(X_test, Y_test, verbose=0)
    print(f"Test Accuracy: {acc:.4f}")

    y_pred = model.predict(X_test).argmax(axis=1)
    print(classification_report(y_test, y_pred))
    return y_pred


def plot_results(history, y_test, y_pred, model):
    fig = plt.figure(figsize=(18, 12))
    gs = gridspec.GridSpec(2, 2)

    ax1 = fig.add_subplot(gs[0, 0])
    ax1.plot(history.history['accuracy'], label='Train')
    ax1.plot(history.history['val_accuracy'], label='Val')
    ax1.legend()
    ax1.set_title("Accuracy")

    ax2 = fig.add_subplot(gs[0, 1])
    ax2.plot(history.history['loss'], label='Train')
    ax2.plot(history.history['val_loss'], label='Val')
    ax2.legend()
    ax2.set_title("Loss")

    ax3 = fig.add_subplot(gs[1, 0])
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', ax=ax3)
    ax3.set_title("Confusion Matrix")

    ax4 = fig.add_subplot(gs[1, 1])
    ax4.axis('off')

    layers_info = []
    total_params = 0

    for layer in model.layers:
        params = layer.count_params()
        total_params += params

        if params > 0:
            try:
                output_shape = str(layer.output.shape)
            except:
                output_shape = "N/A"

            layers_info.append([
                layer.name,
                output_shape,
                f"{params:,}"
            ])

    table = ax4.table(
        cellText=layers_info,
        colLabels=['Layer', 'Output Shape', 'Params'],
        loc='center'
    )
    table.auto_set_font_size(False)
    table.set_fontsize(8)

    ax4.set_title(f"Total Params: {total_params:,}")

    plt.show()


if __name__ == "__main__":
    X_train, X_test, y_train, y_test = load_uci_har()

    X_train, X_test, Y_train, Y_test, y_train, y_test = preprocess(
        X_train, X_test, y_train, y_test
    )

    model = build_cnn_lstm((128, 9), 6)
    model.summary()

    history = train_model(model, X_train, Y_train, X_test, Y_test)

    y_pred = evaluate_model(model, X_test, Y_test, y_test)

    plot_results(history, y_test, y_pred, model)
```
